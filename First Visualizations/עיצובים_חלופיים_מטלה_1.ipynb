```python
import matplotlib.pyplot as plt
import seaborn as sns

top_cities_sorted = top_cities.sort_values(ascending=True)

# Apply a seaborn theme (choose from available themes like 'darkgrid', 'whitegrid', etc.)
xticks_range = np.arange(0, top_cities_sorted.max() + 10, 50)  # Adjust max + 10 to include the highest value
sns.set_theme(style="whitegrid")
top_cities_sorted.plot(kind='barh')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.xticks(xticks_range, rotation=45)
plt.xlabel("Frequency", fontsize=14)
plt.ylabel("City", fontsize=14)
plt.title("Frequency of Incidents by City", fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd

# Assuming Task1Data is a pandas DataFrame with 'city', 'latitude', and 'longitude' columns

# Filter Task1Data for cities in top_cities
cities_to_label = Task1Data[Task1Data['city'].isin(top_cities.index)]

# Create a GeoDataFrame from the filtered data
gdf_cities = gpd.GeoDataFrame(
    cities_to_label[['city', 'latitude', 'longitude']],
    geometry=gpd.points_from_xy(cities_to_label['longitude'], cities_to_label['latitude']),
    crs="EPSG:4326"  # Assuming your coordinates are in WGS84
)

# Load the shapefile for world boundaries (assuming you have it)
world = gpd.read_file("/content/ne_110m_admin_0_countries.shp")

# Filter for Israel only
israel = world[world['NAME'] == "Israel"]  # Replace 'NAME' with the actual column name

# Create the density plot with city labels
fig, ax = plt.subplots(figsize=(20, 20))
israel.plot(ax=ax, color='lightgrey')
sns.kdeplot(data=Task1Data, x="longitude", y="latitude", cmap="Reds", fill=True, ax=ax)

labeled_cities = set()  # Create a set to store labeled cities
texts = []
for i, row in gdf_cities.iterrows():
    city_name = row['city']
    if city_name not in labeled_cities:
        texts.append(plt.text(row['geometry'].x, row['geometry'].y, row['city'], fontsize=15))
        labeled_cities.add(city_name)  # Add labeled city to the set

adjust_text(texts, arrowprops=dict(arrowstyle='->', color='blue'))
plt.title("Incident Density in Israel with City Labels")
plt.show()
```
